**Creating dataset for an And Gate**

```python
x = np.array([[0,0], [0,1], [1,0], [1,1]], "uint8")
y = np.array([[0], [0], [0], [1]], "uint8")
```

**Creating a sequential model using tensorflow**

```python
model = Sequential()
model.add(Dense(units=1, activation='sigmoid', input_dim=2))  ## MODEL CONTAINS ONLY ONE OUTPUT NEURON
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model. fit(x, y, epochs=50)
```

```python
model.fit(x,y,epochs=500)
```

```python
print(model.layers[0].get_weights())
```

```python
model.predict(x)
```

**Deep Neural Network to Predict hand written Digit Recognition**

```python
## importing dataset

from sklearn.datasets import load_breast_cancer
Data = load_breast_cancer()
```

```python
print(Data.DESCR)
```

```python
features = Data.data
target = Data.target

print(features.shape)
print(target.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, input_dim = 30, activation='relu'))  ##hidden layer 1
model.add(Dense(64, activation='relu'))  ##hidden layer 2
model.add(Dense(1, activation='sigmoid')) 
```

```python
model.compile(loss='binary_crossentropy' , optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
scores = model.evaluate(X_test, y_test)
print(scores)  ## returns loss and accuracy
```

```python
predictions = model.predict(X_test)
label=[]
for pred in predictions:
   if pred>=0.5:
       print("Malignent")
   else:
       print("Benign")
```
